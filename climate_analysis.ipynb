```python
import numpy as np
import pandas as pd
```

```python
import datetime as dt
```

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base #
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func #  allow our sql command to create the engine
```

```python
engine = create_engine("postgresql://postgres:postgresql@localhost:5432/Hawaii_db")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
['measurement','station']
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
prev_year = dt.date(2017,8,23) - dt.timedelta(days =365)
prev_year
```

```python

# to get all date and precipitation from measurement
results = session.query(Measurement.date, Measurement.prcp).all()
# create a data frame with  two columns
df=pd.DataFrame(results, columns=['date', 'precipitation'])
# set the date columns as index
df.set_index(df['date'], inplace=True)
df = df.sort_index()
df.drop(columns= ['date'] , inplace= True)
#df.plot(rot=45)
df.head()

```

```python
# plot the dataframe
df.plot(df['date'], df['precipitation'])
plt.xlabel("Date")
plt.ylabel("Rain in Inches")
plt.title("Precipitation Analysis (8/24/16 to 8/23/17)")
plt.legend(["Precipitation"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
